```python
from google.colab import userdata
GEMINI_API_KEY = userdata.get('GOOGLE_API_KEY')
if(GEMINI_API_KEY):
  print("API key found")
else:
  print("API key not found")
```

```python
from google import genai
from google.genai import Client

client = Client = genai.Client(
    api_key=GEMINI_API_KEY,
)
model: str = "gemini-2.0-flash-exp"

```

```python
video_file_name = "/content/Video.mp4"
```

```python
import time

def upload_video(video_file_name):
    # Start uploading the video file
    video_file = client.files.upload(path=video_file_name)

    # Polling for the video processing status
    while video_file.state == "PROCESSING":
        print('Waiting for video to be processed....')
        time.sleep(10)  # Wait for 10 seconds before checking again
        video_file = client.files.get(name=video_file.name or " ")

    # Check if the processing failed
    if video_file.state == "FAILED":
        raise ValueError(f"Video processing failed with state: {video_file.state}")

    # Print the URL of the processed video
    print(f'Video processing complete: {video_file.uri or "No URL available"}')

    return video_file

# Example usage
video_file_name = "/content/non_voice.mp4"  # Replace with your actual video file name
my_video = upload_video(video_file_name)
```

```python
from google.genai.types import Content, Part

# Define the prompt for generating captions
prompt = """For each scene in this video,
            generate captions that describe the scene along with any spoken text placed in quotation marks.
            Place each caption into an object with the timecode of the caption in the video.
            """

# Assume `my_video` is defined and contains the uploaded video information
video = my_video

# Generate content using the model
response = client.models.generate_content(
    model=model,
    contents=[
        Content(
            role="user",
            parts=[  # Corrected 'Parts' to 'parts' (case-sensitive)
                Part.from_uri(
                    file_uri=video.uri or "",
                    mime_type=video.mime_type or ""
                )
            ]
        ),
        prompt,
        # Content(role="user", parts=[Part(text=prompt)])  # Added prompt as a separate Content object
    ]
)

# Print the generated captions
print(response.text)
```
